```python
import wispshapes as shapes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline
```

```python
#use the iris data set
data=sns.load_dataset('iris')
```

choose random rows, create boxes around them, see which boxes overlap

```python
#create two boxes
box1, box2, box3= (shapes.Box(), shapes.Box(),  shapes.Box())
box1.data=data.sample(frac=0.3)
box2.data=data.sample(frac=0.3)
box3.data=data.sample(frac=0.3)
box1.alpha=0.01
box2.alpha=0.1
box2.alpha=0.4
```

```python
#use the selector 
sel=shapes.Selector()
sel.logic='and'
sel.shapes=[box1, box2]
```

visualize the result

```python
fig, ax=plt.subplots()
plt.scatter(sel.data.x, sel.data.y, label='all', c='#01FF70')
#plot the boxe
box1.plot(ax=ax)
box2.plot(ax=ax)
box3.plot(ax=ax)
#plot selected points in different colors
orslct=sel.select(logic='or')
andslct=sel.select(logic='and')
plt.scatter(orslct.x, orslct.y, c='#111111', label='or')
plt.scatter(andslct.x, andslct.y, c='#AAAAAA', label='and')
ax.set_ylim([1.9, 4.0])
plt.legend(loc='best')
```
