# meta4+dacl1k

```python
df1  =pd.read_csv("meta4.csv")
df2  =pd.read_csv("dacl1k.csv")
df2["name"]  ="dacl1k"
```

```python
df1
```

```python
string_list = []
class_names = ['NoDamage', 'Crack', 'Efflorescence', 'Spalling', 'BarsExposed', 'Rust']
for index, row in tqdm(df1.iterrows()):
    string = ""
    for idx, class_name in enumerate(class_names):        
        if row[class_name] > 0: 
           string += class_name[0]
    string_list.append(string)

df1["CLASSES_STRING"] = string_list
df1["CLASSES_N"] = df1["CLASSES_STRING"].apply(lambda x: len(x))
```

```python
df1 = df1[order]
df2 = df2[order]
```

```python
df1
```

```python
df2
```

```python
df1.to_csv("meta4.csv", index=False)
df.to_csv("meta3+dacl1k.csv", index=False)
```
